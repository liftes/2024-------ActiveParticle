```python
def generate_scr_file(txt_file_path, scr_file_path, dx, dy, flag=False):
    try:
        # 读取txt文件
        with open(txt_file_path, 'r') as txt_file:
            lines = txt_file.readlines()

        xlist = []
        ylist = []

        # 打开或创建scr文件
        with open(scr_file_path, 'w') as scr_file:
            for line in lines:
                # 解析每行的xy坐标
                x, y = map(float, line.split())
                if flag:
                    x = (x + 75)%150

                x = x + dx
                y = y + dy

                xlist.append(x)
                ylist.append(y)
                
                # 写入scr文件内容
                scr_file.write("circle\n")
                scr_file.write(f"{x:.1f},{y:.1f}\n")
                scr_file.write("0.89\n")

        print(f"Successfully created {scr_file_path} from {txt_file_path}")
    
    except Exception as e:
        print(f"An error occurred: {e}")

    return xlist, ylist

# 示例用法
txt_file_path = "Locations/lattice621.txt"  # 替换为你的txt文件路径
scr_file_path = "output-六角.scr"  # 替换为你希望生成的scr文件路径
xlist, ylist = generate_scr_file(txt_file_path, scr_file_path, 5, 50, True)
# plt.figure(figsize=(2, 8))
# plt.plot(xlist, ylist, ".")

txt_file_path = "Locations/Hyperuniformity621.txt"  # 替换为你的txt文件路径
scr_file_path = "output-超匀.scr"  # 替换为你希望生成的scr文件路径
xlist, ylist = generate_scr_file(txt_file_path, scr_file_path, 160, 50, False)
# plt.figure(figsize=(2, 8))
# plt.plot(xlist, ylist, ".")

txt_file_path = "Locations/Random670.txt"  # 替换为你的txt文件路径
scr_file_path = "output-随机.scr"  # 替换为你希望生成的scr文件路径
xlist, ylist = generate_scr_file(txt_file_path, scr_file_path, 315, 50, False)
# plt.figure(figsize=(2, 8))
# plt.plot(xlist, ylist, ".")
```

```python
def read_circle_centers_from_scr(scr_file_path):
    centers = []
    try:
        with open(scr_file_path, 'r') as scr_file:
            lines = scr_file.readlines()

            # 遍历文件中的每行，查找圆的圆心坐标
            i = 0
            while i < len(lines):
                if lines[i].strip().lower() == "circle":
                    # 提取圆心坐标
                    center = lines[i+1].strip().split(',')
                    x = float(center[0])
                    y = float(center[1])
                    centers.append((x, y))
                    i += 3  # 跳过半径行
                else:
                    i += 1
        print(f"Successfully read {len(centers)} circle centers from {scr_file_path}")
    except Exception as e:
        print(f"An error occurred: {e}")
    
    return centers

def plot_circle_centers(centers, figsize=(4.5, 6.5)):
    if not centers:
        print("No circle centers to plot.")
        return
    
    x_coords = [c[0] for c in centers]
    y_coords = [c[1] for c in centers]

    plt.figure(figsize=figsize)
    plt.scatter(x_coords, y_coords, c='blue', marker='.')
    plt.xlabel('X coordinate')
    plt.ylabel('Y coordinate')
    plt.title('Scatter Plot of Circle Centers')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    scr_file_path = "output-合并.scr"  # 替换为你的SCR文件路径
    centers = read_circle_centers_from_scr(scr_file_path)
    plot_circle_centers(centers)
```

```python
import numpy as np
import matplotlib.pyplot as plt



def generate_hexagonal_grid(x_min, x_max, y_min, y_max, scr_file_path, a):
    try:
        # 计算六角密排结构的基矢
        dx = a
        dy = np.sqrt(3) * a / 2
        
        # 计算网格中心
        x_center = (x_max + x_min) / 2
        y_center = (y_max + y_min) / 2
        
        # 生成x和y方向上的点
        xlist = []
        ylist = []
        
        x_start = x_min - a
        x_end = x_max + a
        y_start = y_min - dy
        y_end = y_max + dy
        
        for j in np.arange(y_start, y_end, dy):
            for i in np.arange(x_start, x_end, dx):
                x_shift = a / 2 if int(j / dy) % 2 == 1 else 0
                x = i + x_shift
                y = j
                
                if x_min <= x <= x_max and y_min <= y <= y_max:
                    xlist.append(x)
                    ylist.append(y)
        
        # 将结构居中
        x_offset = x_center - np.mean(xlist)
        y_offset = y_center - np.mean(ylist)
        
        xlist = [x + x_offset for x in xlist]
        ylist = [y + y_offset for y in ylist]
        
        # 生成scr文件
        with open(scr_file_path, 'w') as scr_file:
            for x, y in zip(xlist, ylist):
                scr_file.write("circle\n")
                scr_file.write(f"{x:.1f},{y:.1f}\n")
                scr_file.write("0.89\n")
        
        print(f"Successfully created {scr_file_path}")
        
        # 绘制预览图形
        plt.figure(figsize=(3, 3))
        plt.plot(xlist, ylist, 'o')
        plt.xlim([x_min, x_max])
        plt.ylim([y_min, y_max])
        plt.gca().set_aspect('equal', adjustable='box')
        plt.show()
    
    except Exception as e:
        print(f"An error occurred: {e}")

# 示例用法
x_min = 0
x_max = 150
y_min = 0
y_max = 600
scr_file_path = "output_hexagonal_grid.scr"
a = 13  # 晶格常数

generate_hexagonal_grid(x_min, x_max, y_min, y_max, scr_file_path, a)
```

```python
def add_rectangle_to_scr(scr_file_path, rect_top_left, rect_bottom_right):
    try:
        # 打开并追加写入SCR文件
        with open(scr_file_path, 'a') as scr_file:
            # 添加矩形的内容
            scr_file.write("rectangle\n")
            scr_file.write(f"{rect_top_left[0]:.1f},{rect_top_left[1]:.1f}\n")
            scr_file.write(f"{rect_bottom_right[0]:.1f},{rect_bottom_right[1]:.1f}\n")
        
        print(f"Successfully added a rectangle to {scr_file_path}")
    
    except Exception as e:
        print(f"An error occurred: {e}")
    # try:
    #     # 打开并追加写入SCR文件
    #     with open(scr_file_path, 'a') as scr_file:
    #         # 计算矩形四个角的坐标
    #         x1, y1 = rect_top_left
    #         x2, y2 = rect_bottom_right

    #         # 添加四条直线的内容，形成一个矩形
    #         # 上边
    #         scr_file.write("line\n")
    #         scr_file.write(f"{x1:.1f},{y1:.1f}\n")
    #         scr_file.write(f"{x2:.1f},{y1:.1f}\n")
    #         # 右边
    #         scr_file.write("line\n")
    #         scr_file.write(f"{x2:.1f},{y1:.1f}\n")
    #         scr_file.write(f"{x2:.1f},{y2:.1f}\n")
    #         # 下边
    #         scr_file.write("line\n")
    #         scr_file.write(f"{x2:.1f},{y2:.1f}\n")
    #         scr_file.write(f"{x1:.1f},{y2:.1f}\n")
    #         # 左边
    #         scr_file.write("line\n")
    #         scr_file.write(f"{x1:.1f},{y2:.1f}\n")
    #         scr_file.write(f"{x1:.1f},{y1:.1f}\n")

    #     print(f"Successfully added rectangle as boundary lines to {scr_file_path}")
    
    # except Exception as e:
    #     print(f"An error occurred: {e}")

def generate_scr_file(txt_file_path, scr_file_path, section, circle_radius=0.89, board_thickness=4.5, gap_between_boards=10, y_offset=10, board_width=150, flag=False):
    # 确定每个部分的水平偏移
    offset_map = {
        'left': board_thickness + gap_between_boards,  # 第一个板的起始点
        'middle': board_thickness + gap_between_boards + board_width + gap_between_boards + board_thickness + gap_between_boards,  # 第二个板的起始点
        'right': board_thickness + gap_between_boards + 2 * (board_width + gap_between_boards + board_thickness + gap_between_boards)  # 第三个板的起始点
    }
    try:
        # 读取txt文件
        with open(txt_file_path, 'r') as txt_file:
            lines = txt_file.readlines()

        # 打开或创建scr文件
        with open(scr_file_path, 'w') as scr_file:
            # 偏移计算
            dx = offset_map[section]

            for line in lines:
                # 解析每行的xy坐标
                x, y = map(float, line.split())
                
                if flag:
                    x = (x + board_width/2)%board_width

                # 应用水平偏移
                x = x + dx
                y = y + y_offset  # 垂直方向的偏移
                
                # 写入scr文件内容
                scr_file.write("circle\n")
                scr_file.write(f"{x:.1f},{y:.1f}\n")
                scr_file.write(f"{circle_radius}\n")  # 圆的半径

        rect_top_left = (offset_map[section],
                        y_offset-gap_between_boards)  # 矩形左上角坐标
        rect_bottom_right = (offset_map[section]+board_width,           
                            y_offset-gap_between_boards-board_thickness)  # 矩形右下角坐标
        add_rectangle_to_scr(scr_file_path, rect_top_left, rect_bottom_right)
        
        rect_top_left = (offset_map[section],
                        y_offset+gap_between_boards+board_high)  # 矩形左上角坐标
        rect_bottom_right = (offset_map[section]+board_width,
                            y_offset+gap_between_boards+board_high+board_thickness)  # 矩形右下角坐标
        
        add_rectangle_to_scr(scr_file_path, rect_top_left, rect_bottom_right)


        print(f"Successfully created {scr_file_path} from {txt_file_path}")
    
    except Exception as e:
        print(f"An error occurred: {e}")

# 示例用法
circle_radius=0.89
board_thickness=4.5
gap_between_boards=10
y_offset = 80
board_width=150
board_high=600

txt_file_path = "Locations/lattice621.txt"
scr_file_path = "output-六角.scr"
generate_scr_file(txt_file_path, scr_file_path, 'left',
                circle_radius = circle_radius,
                board_thickness = board_thickness,
                gap_between_boards = gap_between_boards,
                y_offset=y_offset,
                flag = True)

txt_file_path = "Locations/Hyperuniformity621.txt"
scr_file_path = "output-超匀.scr"
generate_scr_file(txt_file_path, scr_file_path, 'middle',                
                circle_radius = circle_radius,
                board_thickness = board_thickness,
                gap_between_boards = gap_between_boards,
                y_offset=y_offset)

txt_file_path = "Locations/random670.txt"
scr_file_path = "output-随机.scr"
generate_scr_file(txt_file_path, scr_file_path, 'right',                        
                circle_radius = circle_radius,
                board_thickness = board_thickness,
                gap_between_boards = gap_between_boards,
                y_offset=y_offset)
```

```python
def add_vertical_holes_to_scr(scr_file_path, dy, y0, board_thickness=4.5, hole_length=5):
    # 计算每个矩形孔的宽度和长度
    hole_width = board_thickness
    try:
        # 打开并追加写入SCR文件
        with open(scr_file_path, 'a') as scr_file:
            # 设置起始和终止的y坐标，每个间隔dy添加一个矩形孔
            y = y0
            while y + hole_length <= board_high + 2 * y_offset:
                # 四个板厚度的位置
                positions = [
                    0,  # 最左侧
                    board_width + board_thickness + 2 *gap_between_boards,
                    2*board_width + 2*board_thickness + 4 *gap_between_boards,
                    3*board_width + 3*board_thickness + 6 *gap_between_boards,
                ]
                for x in positions:
                    rect_top_left = (x, y)
                    rect_bottom_right = (x + hole_width, y + hole_length)
                    # 添加矩形孔的内容
                    scr_file.write("rectangle\n")
                    scr_file.write(f"{rect_top_left[0]:.1f},{rect_top_left[1]:.1f}\n")
                    scr_file.write(f"{rect_bottom_right[0]:.1f},{rect_bottom_right[1]:.1f}\n")
                y += dy

        print(f"Successfully added vertical holes to {scr_file_path}")
    
    except Exception as e:
        print(f"An error occurred: {e}")

# 示例用法
scr_file_path = "output-合并.scr"  # 使用已经合并并添加了边框的SCR文件
dy = 100  # 纵向间隔100mm
y0 = y_offset
hole_length = 5
add_vertical_holes_to_scr(scr_file_path, dy, y0, board_thickness, hole_length)
```

```python
def add_horizontal_holes_to_scr(scr_file_path, dx, x0, hole_width=5, board_thickness=4.5):
    try:
        # 打开并追加写入SCR文件
        with open(scr_file_path, 'a') as scr_file:
            # 设置起始和终止的x坐标，每个间隔dx添加一个矩形孔
            x = x0
            total_width = 3*board_width + 4*board_thickness + 6*gap_between_boards
            while x + hole_width <= total_width:
                # 上下边界的位置
                positions = [
                    0,  # 上边界
                    board_high + 2*y_offset - board_thickness       # 下边界
                ]
                for y in positions:
                    rect_top_left = (x, y)
                    rect_bottom_right = (x + hole_width, y + board_thickness)
                    # 添加矩形孔的内容
                    scr_file.write("rectangle\n")
                    scr_file.write(f"{rect_top_left[0]:.1f},{rect_top_left[1]:.1f}\n")
                    scr_file.write(f"{rect_bottom_right[0]:.1f},{rect_bottom_right[1]:.1f}\n")
                    # add_rectangle_to_scr(scr_file_path, rect_top_left, rect_bottom_right)
                x += dx

        print(f"Successfully added horizontal holes to {scr_file_path}")
    
    except Exception as e:
        print(f"An error occurred: {e}")

# 示例用法
scr_file_path = "output-合并.scr"  # 使用已经合并并添加了边框的SCR文件
dx = 100  # 横向间隔100mm
x0 = board_thickness + gap_between_boards  # 从第一个板厚度后开始
add_horizontal_holes_to_scr(scr_file_path, dx, x0, hole_length, board_thickness)
```
